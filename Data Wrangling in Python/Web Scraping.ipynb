```python
all_years = []
for i in range(2010,2022):
    cfb_defense = 'http://www.cfbstats.com/' + str(i) + '/leader/national/team/defense/split01/category10/sort01.html'
    all_years.append(cfb_defense)
print(all_years)
# just checking that I'm generating the url's I need to collect each year's table
```

```python
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# scrape the tables for all years (2010-2021) with each year an element of a list
html_list =[]
for year in all_years:
    page = driver.get(year)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    html_list.append(soup)
```

```python
# taking a look at one of the html files to examin the structure
# I am expecting the structure of the file for each year to be the same
print(html_list[0].prettify())
```

```python
for year in html_list:
    print(year.title.text)
# examine the title of each page
```

```python
# examine the links on each page
for year in html_list:
    for link in year.find_all("a"):
        print("Inner Text: {}".format(link.text))
        print("Title: {}".format(link.get("title")))
        print("href: {}".format(link.get("href")))
```

```python
# examine the tabless and rows in each year
years=[]
for year in html_list:
    leaders_table = year.find("table", attrs={"class":"leaders"})
    years.append(leaders_table.tbody.find_all("tr"))
for year in years:
    print(year[1:])
```

```python
# get the table headers 
t_header=[]
t_header.append(years[0][0].find("th", attrs={"class":"rank"}))
t_header.append(years[0][0].find("th", attrs={"class":"team-name"}))
t_header.append(years[0][0].find("th", attrs={"class":"g"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-rush"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-pass"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-play"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-yard"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-yard-play"}))
t_header.append(years[0][0].find("th", attrs={"class":"total-off-yard-g highlight"}))

# extract the text from each header, and add header for team rank (the rank column in the html file has no header)
# note that the table headers are the same for each year, so I was able to pull them from just one of the tables
t_headers=[]
for header in t_header:
    t_headers.append(header.text)
t_headers[0] = 'Rank'
t_headers
```

```python
# create a list of dictionaries, one for each row of the table for a given year
year_data=[]
for year in years:
    table_data=[]
#     the first element in year is the table headers, the data starts with the second element
    for tr in year[1:]:
        t_row={}
#         iterate over the table rows and tie to the table headers in a dictionary
        for td, th in zip(tr.find_all("td"), t_headers):
            t_row[th]=td.text.strip()
#             add each dictionary to the list
        table_data.append(t_row)
#     add each year's list of dictionaries to a list of all the years
    year_data.append(table_data)
    
#     below I check the total number of rows across all years, which I will use a check for my final df
num=[]
for year in year_data:
    num.append(len(year))
sum(num)
```

```python
# create a dataframe from each dictionary, pivot, and concatenate to a dataframe for one year
# NOTE: I have to believe that there is a better way to do this than to create a dataframe out of each row
# and concatenate all the row dataframes into a dataframe for a given year, but after spending a few hours
# trying to do this more efficiently, I decided to go with what works.  Any guidance you could provide would
# be much apprecited

import pandas as pd

ranks=[]
# iterate on i so that I can include a year column in each year dataframe
for i in range(0, len(year_data)):
    dataframes = []
    for dct in year_data[i]:
#         generate a dataframe from each dictionary
        df = pd.DataFrame(list(dct.items()))
#         pivot to make the first columne the headers and the second column the row
        df = pd.pivot_table(df,columns=0, aggfunc = 'sum')
#     add to the list of dataframes (one dataframe for each row of the year table)
        dataframes.append(df)
# for a given year, concatenate all of the row dataframes into a single dataframe for the year
    df2 = pd.concat(dataframes).reset_index()
    df2.drop('index', axis = 1, inplace=True)
#     add a column for year
    df2['Year'] = 2010+i
# rearrange columns
    new_cols = ['Year', 'Rank', 'Name', 'G', 'Rush Yards', 'Pass Yards', 'Total Yards', 'Plays', 
        'Yards/G', 'Yards/Play']
    rank = df2[new_cols]
# add the dataframe for each year to a list of dataframes
    ranks.append(rank)
    
#     check that I have 12 dataframes, one for each year
print(len(ranks))
```
